```python
import pandas as pd
pd.options.display.max_columns = None
```

```python
disaster_data = pd.read_excel("data\disaster_loss_data_indonesia.xlsx")
disaster_data.head()
```

```python
df_data_bencana_1 = pd.read_excel("data/Data Bencana Indonesia 2014-2019.xlsx")
df_data_bencana_1.head()
```

```python
df_data_bencana_2 = pd.read_excel("data/Data Bencana Indonesia 2019-2023.xlsx")
df_data_bencana_2.head()
```

```python
data_bencana = pd.concat([df_data_bencana_1, df_data_bencana_2])
data_bencana
```

```python
data_bencana = data_bencana.sort_values(by=["Tanggal Kejadian"], ascending=True)
data_bencana = data_bencana.drop(columns=['No.'])
data_bencana = data_bencana.drop(columns=['Kode Identitas Bencana'])
data_bencana.head()
```

```python
data_bencana.columns.tolist()
```

```python
data_bencana = data_bencana.rename(columns={
                                            'ID Kabupaten'              : 'Code Province',
                                             'Tanggal Kejadian'         : 'Date',
                                             'Kejadian'                 : 'Event',
                                             'Lokasi'                   : 'Location',
                                             'Kabupaten'                : 'District',
                                             'Provinsi'                 : 'Province',
                                             'Kronologi & Dokumentasi'  : 'Description of Cause',
                                             'Penyebab'                 : 'Cause',
                                             'Meninggal'                : 'Deaths',
                                             'Hilang'                   : 'Missing',
                                             'Terluka'                  : 'Injured',
                                             'Rumah Rusak'              : 'Houses Damaged',
                                             'Rumah Terendam'           : 'Houses Submerged',
                                             'Fasum Rusak'              : 'Public Facilities Damaged',
                                            })
```

```python
data_bencana.head()
```

```python
sorted(data_bencana['Event'].unique().tolist())
```

```python
sorted(disaster_data['Event'].unique().tolist())
```

```python
data_bencana.loc[data_bencana['Event'] ==  'BANJIR'                    , ['Event']] = ['FLOODS']
data_bencana.loc[data_bencana['Event'] ==  'GELOMBANG PASANG / ABRASI' , ['Event']] = ['SURGE']
data_bencana.loc[data_bencana['Event'] ==  'GEMPA BUMI'                , ['Event']] = ['EARTHQUAKE']
data_bencana.loc[data_bencana['Event'] ==  'GEMPA BUMI DAN TSUNAMI'    , ['Event']] = ['EARTHQUAKE AND TSUNAMI']
data_bencana.loc[data_bencana['Event'] ==  'KEBAKARAN HUTAN DAN LAHAN' , ['Event']] = ['FOREST FIRE']
data_bencana.loc[data_bencana['Event'] ==  'KEKERINGAN'                , ['Event']] = ['DROUGHT']
data_bencana.loc[data_bencana['Event'] ==  'LETUSAN GUNUNG API'        , ['Event']] = ['ERUPTION']
data_bencana.loc[data_bencana['Event'] ==  'PUTING BELIUNG'            , ['Event']] = ['STRONG WIND']
data_bencana.loc[data_bencana['Event'] ==  'TANAH LONGSOR'             , ['Event']] = ['LANDSLIDES']
data_bencana.loc[data_bencana['Event'] ==  'TSUNAMI'                   , ['Event']] = ['TSUNAMI']
```
